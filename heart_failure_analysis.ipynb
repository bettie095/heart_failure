```python
#to save the html table as pandas data frame
column_details_df = pd.read_html("https://bmcmedinformdecismak.biomedcentral.com/articles/10.1186/s12911-020-1023-5/tables/1")[0]
#to drop unnecessary column
column_details_df.drop('Range', axis = 1, inplace = True, errors = 'ignore')
#to drop unnecessary rows
column_details_df.drop([3, 4, 7, 13], axis = 0, inplace = True, errors = 'ignore')
#to rename the column name
column_details_df.columns = ['feature', 'explanation', 'measurement_unit']
column_details_df
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
#to display plots, use %matplotlib inline
%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['figure.figsize'] = (10, 6)
matplotlib.rcParams['figure.facecolor'] = 'white'
```

```python
#to display the age distribution
plt.hist(heart_failure_df.age, bins = np.arange(40, 100, 5), color = 'skyblue')
plt.xlabel('Age of Patients(Years)')
plt.ylabel('Number of Patients')
plt.title('Age Distribution')
```

```python
gender_count = heart_failure_df.sex.value_counts()
gender_count
```

```python
plt.figure(figsize = (12, 5))
plt.pie(gender_count, labels = gender_count.index, autopct = '%.1f%%', startangle = 90, explode = [0.1, 0], colors = ['plum', 'mediumpurple'])
plt.title('Gender Distribution');
```

```python
#to calculte death rate
death_count = heart_failure_df.patient_dead.value_counts()
death_count
```

```python
plt.figure(figsize = (12, 6))
plt.pie(death_count, labels = ['Alive', 'Dead'], autopct = '%.1f%%', startangle = 180, colors = ['lightcoral', 'mediumpurple'])
plt.title('Percentage of Patients Dead and Alive');
```
